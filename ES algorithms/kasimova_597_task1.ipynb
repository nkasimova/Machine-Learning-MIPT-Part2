## 1) Realize Winters Model with Linear trend (20%)

```python
# функция WintersLinearTrendExponentialSmoothing была перенесена в модуль utils, иначе buildforecast ее бы не увидел.
# Но для удобства копия кода приведена ниже. 
"""
# Start with this code
###################### Winters Exponential Smoothing #########################
# x <array Tx1>- time series, 
# h <scalar> - forecasting delay
# Params <dict> - dictionary with 
#    alpha <scalar in [0,1]> - smoothing parameter of level
#    beta <scalar in [0,1]> - smoothing parameter of trend
#    gamma <scalar in [0,1]> - smoothing parameter of seasonality
def WintersLinearTrendExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    beta = Params['beta']
    gamma = Params['gamma']
    
    p = Params['seasonality_period']
    
    FORECAST = [np.NaN]*(T+h)## put foreast of the model here
    
    if alpha > 1:
        w.warn('Alpha can not be more than 1')
        #alpha = 1
        return FORECAST
    if alpha < 0:
        w.warn('Alpha can not be less than 0')
        #alpha = 0
        return FORECAST
    
    if beta > 1:
        w.warn('beta can not be more than 1')
        #beta = 1
        return FORECAST
    if beta < 0:
        w.warn('beta can not be less than 0')
        #beta = 0
        return FORECAST
    
    if gamma > 1:
        w.warn('beta can not be more than 1')
        #gamma = 1
        return FORECAST
    if gamma < 0:
        w.warn('beta can not be less than 0')
        #gamma = 0
        return FORECAST
    
    l = np.NaN # initialize ts level 
    b = np.NaN # initialize ts trend
    s = [] # initalize seasonality values (it must be vector of lenth p)
    
    for i in range(p):
        if not math.isnan(x[i]):
            s.append(x[i])
        else:
            s.append(s[i-1])
    
    for t in range(T):
        if not math.isnan(x[t]):
            if math.isnan(l):
                l = 1 # initialize
            if math.isnan(b):
                b = 0
            
            l_old = l
                
            if t < p:
                l = alpha * x[t] / s[t] + (1 - alpha) * (l + b)
                b = beta * (l - l_old) + (1 - beta) * b
            else:
                s_old = s[t % p]
                l = alpha * x[t] / s_old + (1 - alpha) * (l + b) # recurrent smoothing of level 
                b = beta * (l - l_old) + (1 - beta) * b
                s[t % p] = gamma * x[t] / l + (1 - gamma) * s[t % p]
                        
        FORECAST[t + h] = (l + b * h) * s[(t + h) % p]
    return FORECAST
"""
```

**Вывод:**

Посмотрим для начала, как определены критерии в библиотеке dm_test:

MSE : d = (e1)^2 - (e2)^2

MAD : d = abs(e1) - abs(e2)

Poly: d = (e1)^3 - (e2)^3, в данном случае power=3,

где, как сказано в документации к dm_test, e1 - этот разница между предсказанием модели theil-wage и реальными значениями, а e2 - разница между предсказанием модели winters with linear trend и реальными значениями.
